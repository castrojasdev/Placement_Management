```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


companies = pd.read_csv('Company.csv')
resumes = pd.read_csv('Resume.csv')


vectorizer = CountVectorizer(tokenizer=lambda x: x.split(','), lowercase=False)


company_skills = vectorizer.fit_transform(companies['Skills'].fillna(''))


resume_skills = vectorizer.transform(resumes['Skills'].fillna(''))


cosine_similarities = cosine_similarity(resume_skills, company_skills)


for i, row in resumes.iterrows():
    print(f"{row['Name']} - {companies['Company'][cosine_similarities[i].argmax()]}")
```

```python

company_candidates = {}


for i, row in resumes.iterrows():
    company_index = cosine_similarities[i].argmax()
    company_name = companies['Company'][company_index]
    candidate_name = row['Name']
    

    if company_name in company_candidates:
        company_candidates[company_name].append(candidate_name)
    else:
        company_candidates[company_name] = [candidate_name]


for company_name, candidates in company_candidates.items():
    print(f"{company_name} ==> {', '.join(candidates)}")
```
